```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('netflix.csv')
```

```python
movies.head(10)
```

```python
movies.describe()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies=movies[['show_id','title', 'country' , 'description']]
```

```python
movies
```

```python
movies['tags'] = movies['country']+movies['description']
```

```python
movies
```

```python
new_data  = movies.drop(columns=['description','country'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="Jaguar"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommand(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand('Jeans')
```
